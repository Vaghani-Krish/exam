```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data={
    'Student':['A','B','C','D','E','F','G','H'],
    'Marks':[45,55,60,70,75,85,90,100],
    'Attendance':[60,65,70,75,80,85,90,95]
}

df=pd.DataFrame(data)
df
```

```python
freq=df['Marks'].value_counts()
freq
```

```python
bins=[0,20,40,60,80,100]
labels=['0-20','21-40','41-60','61-80','81-100']
df['Class Interval']=pd.cut(df['Marks'],bins=bins,labels=labels)
df
```

```python
grouped_freq=df['Class Interval'].value_counts().sort_index()
grouped_freq
```

```python
plt.hist(df['Marks'],bins=5)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram of Marks')
plt.show()
```

```python
plt.scatter(df['Attendance'],df['Marks'])
plt.xlabel('Attendance(%)')
plt.ylabel('Marks')
plt.title('Scatter Plot:Attendace vs Marks')
plt.show()
```

```python
correction=df[['Marks','Attendance']].corr()
correction
```

```python
plt.boxplot(df['Marks'])
plt.title('Box PLot of Marks')
plt.ylabel('Marks')
plt.show()
```

```python
# normalization

import pandas as pd
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Marks': [45, 55, 60, 70, 90],
    'Attendance': [60, 65, 70, 80, 95]
}
df = pd.DataFrame(data)
df
```

```python
df_norm=df.copy()
```

```python

# min-max
df_norm['Marks']=(df['Marks']-df['Marks'].min())/(df['Marks'].max()-df['Marks'].min())

df_norm['Attendance']=(df['Attendance']-df['Attendance'].min())/(df['Attendance'].max()-df['Attendance'].min())

df_norm
```

```python
# z -score norm.

df_std=df.copy()
```

```python
df_std['Marks']=(df['Marks']-df['Marks'].mean())/df['Marks'].std()

df_std['Attendance']=(df['Attendance']-df['Attendance'].mean())/df['Attendance'].std()

df_std
```

```python
# normalized multiple columns

cols=['Marks','Attendance']

df[cols]=(df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())

df
```

```python
import pandas as pd

data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math', 'Science', 'Math', 'Science'],
    'Marks': [78, 85, 62, 70, 90, 88, 75, 80],
    'Hours_Studied': [5, 6, 4, 5, 7, 6, 5, 6]
}

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df.groupby('Subject')['Marks'].mean()
```

```python
df.groupby('Subject')['Marks'].agg(['mean','min','max','count'])
```

```python
df.groupby(['Subject','Hours_Studied'])['Marks'].mean()
```

```python
df['Marks_per_Hour']=df['Marks']/df['Hours_Studied']
```

```python
df.groupby('Subject')['Marks_per_Hour'].mean()
```

```python
df.groupby('Subject').agg(
    avg_marks=('Marks','mean'),
    max_marks=('Marks','max'),
    avg_hours=('Hours_Studied','mean'),
)
```
