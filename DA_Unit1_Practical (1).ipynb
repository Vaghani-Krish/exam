## 4. Frequency Distribution

```python
# Create Sample Dataset
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Marks': [45, 55, 60, 70, 75, 85, 90, 100],
    'Attendance': [60, 65, 70, 75, 80, 85, 90, 95]
}

df = pd.DataFrame(data)
df
```

## 8. Handling Missing Values

```python
# Dataset with Missing Values
data = {
    'Name': ['Amit', 'Neha', 'Raj', 'Neha', 'Pooja'],
    'Age': [25, 30, np.nan, 30, 28],
    'Marks': [85, 90, 88, 90, np.nan]
}

df = pd.DataFrame(data)
df
```

```python
# Dataset Info and Statistics
print('=== df.info() ===')
df.info()
print('\n=== df.describe() ===')
df.describe()
```

```python
# Check Missing Values
print('Null Values:')
print(df.isnull())
print('\nMissing Value Count:')
print(df.isnull().sum())
```

```python
# Fill Missing Values with Mean
df_filled = df.copy()
df_filled['Age'].fillna(df_filled['Age'].mean(), inplace=True)
df_filled['Marks'].fillna(df_filled['Marks'].mean(), inplace=True)
print('After filling missing values with mean:')
df_filled
```

```python
# Drop Missing Values
df_dropped = df.copy()
df_dropped.dropna(inplace=True)
print('After dropping missing values:')
df_dropped
```

## 9. Handling Duplicate Values

```python
# Dataset with Duplicates
data = {
    'Student': ['A', 'B', 'C', 'A', 'D', 'B', 'E'],
    'Marks': [45, 55, 60, 45, 70, 55, 80],
    'Attendance': [60, 65, 70, 60, 75, 65, 85]
}

df = pd.DataFrame(data)
df
```

```python
# Check for Duplicates
print('All duplicate rows:')
print(df.duplicated())

print('\nDuplicate by Student column:')
print(df.duplicated(subset=['Student']))
```

```python
# Remove duplicate rows (keep first)
df_no_duplicates = df.drop_duplicates()
print('After removing duplicate rows (keep first):')
df_no_duplicates
```

```python
# Remove duplicates based on Student column (keep first)
df_unique_students = df.drop_duplicates(subset=['Student'])
print('Unique students (keep first):')
display(df_unique_students)

# Keep the LAST occurrence
df_last = df.drop_duplicates(subset=['Student'], keep='last')
print('Unique students (keep last):')
display(df_last)

# Remove ALL duplicates (keep none)
df_no_any_duplicates = df.drop_duplicates(subset=['Student'], keep=False)
print('Remove all duplicate students:')
display(df_no_any_duplicates)
```

```python
# Handle duplicates using aggregation
df_avg = df.groupby('Student', as_index=False).mean()
print('Average values per student:')
display(df_avg)

# Count occurrences of each student
print('\nValue counts of Student:')
df['Student'].value_counts()
```

## 10. Data Normalization

```python
# Dataset for Normalization
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Marks': [45, 55, 60, 70, 90],
    'Attendance': [60, 65, 70, 80, 95]
}

df = pd.DataFrame(data)
df
```

```python
# Min-Max Normalization
# Formula: (x - min) / (max - min)
df_norm = df.copy()
df_norm['Marks'] = (df['Marks'] - df['Marks'].min()) / (df['Marks'].max() - df['Marks'].min())
df_norm['Attendance'] = (df['Attendance'] - df['Attendance'].min()) / (df['Attendance'].max() - df['Attendance'].min())

print('After Min-Max Normalization (0 to 1):')
df_norm
```

```python
# Normalize Multiple Columns at Once
df_multi = df.copy()
cols = ['Marks', 'Attendance']
df_multi[cols] = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print('Normalize multiple columns at once:')
df_multi
```

```python
# Z-Score Normalization (Standardization)
# Formula: (x - mean) / std
df_std = df.copy()
df_std['Marks'] = (df['Marks'] - df['Marks'].mean()) / df['Marks'].std()
df_std['Attendance'] = (df['Attendance'] - df['Attendance'].mean()) / df['Attendance'].std()

print('After Z-Score Normalization (mean ≈ 0, std ≈ 1):')
df_std
```

## 11. Descriptive Statistics - Mean, Median, Mode, Variance, Std Dev

```python
# Dataset
data = {'Marks': [45, 55, 60, 70, 75, 85, 90, 100]}
df = pd.DataFrame(data)
df
```

## 12. Correlation and Covariance

```python
# Dataset
data = {
    'Marks': [45, 55, 60, 70, 75, 85, 90, 100],
    'Hours_Studied': [2, 3, 4, 5, 6, 7, 8, 9]
}
df = pd.DataFrame(data)
df
```

```python
# Covariance
# Positive covariance → as hours studied increase, marks increase
covariance_matrix = df.cov()
print('Covariance Matrix:')
display(covariance_matrix)

cov_marks_hours = df['Marks'].cov(df['Hours_Studied'])
print('Covariance (Marks & Hours_Studied):', cov_marks_hours)
```

```python
# Correlation
# Value ranges from -1 to +1
# +1 → perfect positive | -1 → perfect negative | 0 → no relationship
correlation_matrix = df.corr()
print('Correlation Matrix:')
display(correlation_matrix)

corr_marks_hours = df['Marks'].corr(df['Hours_Studied'])
print('Correlation (Marks & Hours_Studied):', corr_marks_hours)
```

```python
# NumPy Method for Covariance and Correlation
print('Covariance Matrix (NumPy):')
print(np.cov(df['Marks'], df['Hours_Studied']))

print('\nCorrelation Matrix (NumPy):')
print(np.corrcoef(df['Marks'], df['Hours_Studied']))
```

## 13. GroupBy - Data Aggregation

```python
# Dataset
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math', 'Science', 'Math', 'Science'],
    'Marks': [78, 85, 62, 70, 90, 88, 75, 80],
    'Hours_Studied': [5, 6, 4, 5, 7, 6, 5, 6]
}
df = pd.DataFrame(data)
df
```

```python
# Dataset Overview
df.describe()
```

```python
# GroupBy: Average Marks per Subject
# Split → Apply → Combine
avg_by_subject = df.groupby('Subject')['Marks'].mean()
print('Average Marks per Subject:')
avg_by_subject
```

```python
# Multiple Aggregations
df.groupby('Subject')['Marks'].agg(['mean', 'min', 'max', 'count'])
```

```python
# Group by Multiple Columns
df.groupby(['Subject', 'Hours_Studied'])['Marks'].mean()
```

```python
# Marks per Hour Efficiency
df['Marks_per_Hour'] = df['Marks'] / df['Hours_Studied']
print('Which subject is more efficient (marks per hour)?')
df.groupby('Subject')['Marks_per_Hour'].mean()
```

```python
# GroupBy with Custom Column Names
df.groupby('Subject').agg(
    Avg_Marks=('Marks', 'mean'),
    Max_Marks=('Marks', 'max'),
    Avg_Hours=('Hours_Studied', 'mean')
)
```

```python
# reset_index() - Convert grouped result back to clean DataFrame
summary = df.groupby('Subject')['Marks'].mean().reset_index()
summary
```
